**1.Группы**

Проверьте, является ли множество целых чисел $Z$ с операцией сложения группой. Если да, является ли она абелевой? Ответ обоснуйте.

**Решение:**

Множество целых чисел $Z$ с операцией сложения является группой, если:

1. Операция сложения является замкнутой, т.е. для любых элементов $a, b \in Z$ результат применения операции к этим элементам ($a + b$) должен также принадлежать множеству $Z$. Данное условие выполняется, т.к. в результате сложения двух целых чисел всегда получается целое число.

2. Операция сложения является ассоциативной, т.е. для любых элементов $a, b, c \in Z
: a + (b + c) = (a + b) + c$. Данное условие выполняется, т.к. это является одним из основных свойств сложения целых чисел (результат сложения не зависит от расстановки скобок).

3. В $Z$ существует нейтральный элемент $e$, такой что $a + e = e + a = a$ для каждого $a \in Z$. Данное условие выполняется, нейтральным элементом для операции сложения является $0$, $a + 0 = 0 + a = a$ для каждого $a \in Z$.

4. Для любого $a \in Z$ существует элемент $-a \in Z$, такой что $a + (-a) = (-a) + a = e$. Данное условие выполняется, т.к. $a + (-a) = (-a) + a = 0$, где $0 = e$ - нейтральный элемент.

Таким образом, множество целых чисел $Z$ с операцией сложения является группой.

Группа называется абелевой, если операция группы коммутативна, т.е. для любых элементов $a, b \in Z: a + b = b + a$. Операция сложения целых чисел является коммутативной, поэтому множество целых чисел $Z$ с операцией сложения является абелевой группой.

**Ответ:** Таким образом, множество целых чисел $Z$ с операцией сложения является абелевой группой.

**2.Кольца и поля**

Образует ли множество квадратных матриц размера
$2 × 2$ с операциями сложения и умножения кольцо? Является ли это кольцо полем? Ответ поясните.

**Решение:**

Кольцом называется непустое множество $K$ с операциями сложения и  умножения, которые обладают следующими свойствами:

1. относительно сложения $K$ есть абелева группа;

2. ассоциативность умножения, т.е. $(AB)C = A(BC)$ для любых $A, B, C \in K$;

3. дистрибутивность умножения относительно сложения, т.е. $A(B + C) = AB + AC$ и $(A + B)C = AC + BC$ для любых $A, B, C \in K$.

Проверим, образует ли множество квадратных матриц размера $2 × 2$ с операциями сложения и умножения кольцо.

Первое условие выполняется, множество квадратных матриц размера $2 × 2$ образует абелеву группу по сложению, т.к.:

*   сложение матриц $2 × 2$ является замкнутой операцией, т.к. в результате сложения всегда получается матрица $2 × 2$;
*   сложение матриц $2 × 2$ является коммутативной операцией, т.е. $A + B = B + A$;
*   сложение матриц $2 × 2$ является ассоциативной операцией, т.е. $A + (B + C) = (A + B) + C$, результат не зависит от расстановки скобок;
*   в $K$ существует нулевой (нейтральный) элемент 0 (нулевая матрица), такой что $A + 0 = A$ для любой матрицы $A$;
*   для каждой матрицы $A$ существует противоположная матрица $-A$, такая что $A + (-A) = 0$.

Второе условие выполняется, умножение матриц размера $2 × 2$ ассоциативно, т.е. $(AB)C = A(BC)$, а также умножение матриц $2 × 2$ является замкнутой операцией, т.к. в результате умножения получается матрица $2 × 2$.

Третье условие также выполняется, для матриц размера $2 × 2$ умножение дистрибутивно относительно сложения:

*   $A(B + C) = AB + AC$;
*   $(A + B)C = AC + BC$.

Таким образом, можно сделать вывод, что множество квадратных матриц размера $2 × 2$ с операциями сложения и умножения образуют кольцо.

Полем называется коммутативное кольцо, в котором ноль отличен от единицы $0_K \neq 1_K$, и всякий ненулевой элемент является обратимым элементом кольца.

Множество квадратных матриц размера $2 × 2$ с операциями сложения и умножения является некоммутативным кольцом.

Например:

$
\begin{pmatrix}
0 & 1 \\
0 & 0
\end{pmatrix}
⋅
\begin{pmatrix}
1 & 0 \\
0 & 0
\end{pmatrix}=
\begin{pmatrix}
0 & 0 \\
0 & 0
\end{pmatrix}
$

$
\begin{pmatrix}
1 & 0 \\
0 & 0
\end{pmatrix}
⋅
\begin{pmatrix}
0 & 1 \\
0 & 0
\end{pmatrix}=
\begin{pmatrix}
0 & 1 \\
0 & 0
\end{pmatrix}
$

Кроме того, не для каждой матрицы $A$ существует обратная матрица $A^{-1}$. Например, если определитель матрицы равен нулю (det$(A) = 0)$, то матрица является вырожденной и для нее не существует обратной матрицы.

**Ответ:** Множество квадратных матриц размера
$2 × 2$ с операциями сложения и умножения образует кольцо, но это кольцо не является полем.

**3.Полиномы как кольцо**

Проверьте, является ли множество полиномов степени не выше $2$ с коэффициентами из $R$ и обычными операциями сложения и умножения кольцом. Укажите его свойства.

**Решение:**

Полином (многочлен) представляет собой сумму одночленов. Множество полиномов степени не выше $2$ с коэффициентами из $R$ можно представить как выражение вида:

$f(x) = a_0 + a_1x + a_2x^2$, где $a_0, a_1, a_2 \in R$

Множество полиномов степени не выше $2$ с коэффициентами из $R$ и обычными операциями сложения и умножения является кольцом, если:

1. относительно сложения есть абелева группа;

2. ассоциативность умножения;

3. дистрибутивность умножения относительно сложения.

Проверим, выполняются ли данные условия.

Первое условие выполняется, множество полиномов степени не выше $2$ с коэффициентами из $R$ и обычными операциями сложения и умножения образует абелеву группу по сложению:

*   сложение полиномов степени не выше $2$ является замкнутой операцией, т.к. в результате сложения получается также полином степени не выше $2$:
$$ \sum_{i=0}^{2} a_ix^i + \sum_{i=0}^{2} b_ix^i = \sum_{i=0}^{2} (a_i + b_i)x^i; $$

*   сложение полиномов степени не выше $2$ является коммутативной операцией:
$$f(x) + g(x) = \sum_{i=0}^{2} (a_i + b_i)x^i = \sum_{i=0}^{2} (b_i + a_i)x^i = g(x) + f(x); $$

*   сложение полиномов степени не выше $2$ является ассоциативной операцией:
$$ (f(x) + g(x)) + v(x) =  \sum_{i=0}^{2} [(a_i + b_i)+ c_i]x^i = \sum_{i=0}^{2} [a_i + (b_i + c_i)]x^i = f(x) + (g(x) + v(x));$$

*   нулевым элементом является нулевой полином, такой что $f(x) + 0 = f(x);$

*   противополжным для полинома $f = \sum_{i=0}^{2} a_ix^i$ будет полином: $ -f = \sum_{i=0}^{2} (-a_i)x^i$.

Умножение полиномов степени не выше $2$ ассоциативно, но при этом не является замкнутой операцией:

$$ \sum_{i=0}^{2}a_ix^i \cdot  \sum_{j=0}^{2}b_ix^j = \sum_{k=0}^{4}c_kx^k, c_k = \sum_{i+j=k}a_ib_j.$$

Так, результат умножения полиномов степени $2$ имеет степень $4$ и не принадлежит множеству полиномов степени не выше $2$.

Третье условие выполняется, умножение полиномов степени $2$ дистрибутивно относительно сложения:

$f(x)(g(x) + v(x)) = f(x)g(x) + f(x)v(x)$

**Ответ:** Множество полиномов степени не выше $2$ с коэффициентами из $R$ и обычными операциями сложения и умножения не является кольцом, т.к. не выполняется условие замкнутости относительно умножения.

**7.Алгоритм Штрассена**

Умножьте матрицы методом Штрассена:

$
F = \begin{pmatrix}
1 & 2 \\
3 & 4
\end{pmatrix},
G = \begin{pmatrix}
5 & 6 \\
7 & 8
\end{pmatrix} $

Опишите шаги разбиения на блоки и промежуточные вычисления.

**Решение:**

Алгоритм Штрассена предполагает сведение матриц к специальному блочному виду. Так, для умножения матриц $2$ x $2$ требуется 7 рекурсивных умножений и 18 сложений.

Пусть:

$
F = \begin{pmatrix}
F_{11} & F_{12} \\
F_{21} & F_{22}
\end{pmatrix},
G = \begin{pmatrix}
G_{11} & G_{12} \\
G_{21} & G_{22}
\end{pmatrix} $

$P_1 = (F_{11} + F_{22})(G_{11} + G_{22}) = (1 + 4) \cdot (5 + 8) = 65 $

$P_2 = (F_{21} + F_{22})G_{11} = (3 + 4) \cdot 5 = 35$

$P_3 = F_{11}(G_{12} - G_{22}) = 1 \cdot (6 - 8) = -2$

$P_4 = F_{22}(G_{21} - G_{11}) = 4 \cdot (7 - 5) = 8$

$P_5 = (F_{11} + F_{12})G_{22} = (1 + 2) \cdot 8 = 24$

$P_6 = (F_{21} - F_{11})(G_{11} + G_{12}) = (3 - 1) \cdot (5 + 6) = 22$

$P_7 = (F_{12} - F_{22})(G_{21} + G_{22}) = (2 - 4) \cdot (7 + 8) = -30$

$
F \cdot G =
\begin{pmatrix}
P_1 + P_4 - P_5 + P_7 & P_3 + P_5 \\
P_2 + P_4 & P_1 - P_2 + P_3 + P_6
\end{pmatrix} =
\begin{pmatrix}
65 + 8 - 24 + (-30) & -2 + 24 \\
35 + 8 & 65 - 35 + (-2) + 22
\end{pmatrix} =
\begin{pmatrix}
19 & 22 \\
43 & 50
\end{pmatrix}
$

**Ответ:** $ F \cdot G =
\begin{pmatrix}
19 & 22 \\
43 & 50
\end{pmatrix}
$

**8.Сравнение сложности**

Сравните количество арифметических операций для умножения матриц $4 × 4$ стандартным методом и методом Штрассена. Приведите формулы расчёта.

**Решение:**

Стандартный метод умножения матриц предполагает, что каждый элемент строки первой матрицы умножается на соответствующий элемент столбца второй матрицы, а затем суммируется. Для умножения матриц $n × n$ требуется $n^3$ отдельных умножений, а для сложения $n^2(n-1)$.

Так, для умножения матриц $4 × 4$ стандартным методом потребуется $4^3$, или $64$, операции умножения, а также $4^2 \cdot (4-1) = 48$ операций сложения.

Таким образом, общее количество операций для умножения матриц $4 × 4$ стандартным методом составляет:

$64$ операции умножения $+$ $48$ операций сложения $= 112$

Метод Штрассена же предполагает, что матрицы размером $n × n$ можно умножить меньше, чем за $n^3$ отдельных умножений. В основе данного метода находится процесс декомпозиции, то есть разбиение матрицы на блоки - меньшие подматрицы. Для умножения матриц требуется $7$ рекурсивных операций умножения и $18$ операций сложения и вычитания (на каждом уровне рекурсии).

Так, для умножения матриц $4 × 4$ методом Штрассена необходимо разбить матрицы на блоки $2 × 2$. Для умножения матриц $2 × 2$ требуется $2^3$, или $8$, операций умножения, а также $2^2(2-1)$, или $4$, операции сложения. Так как для умножения матриц методом Штрассена необходимо выполнить 7 рекурсивных операций умножения, то получаем: $7 \cdot 8 = 56$ операций умножения и $7 \cdot 4 = 28$ операций сложения. Также необходимо выполнить $18$ операций сложения и вычитания матриц $2 × 2$, каждое сложение двух матриц $2 × 2$ требует $4$ скалярных сложений, отсюда необходимо:
$4 \cdot 18 = 72$. Так, общее число операций сложения и вычитания составляет: $28 + 72 = 100$

Таким образом, общее количество операций для умножения матриц $4 × 4$ методом Штрассена составляет:

$56$ операций умножения $+$ $100$ операций сложения и вычитания $= 156$

Как можно отметить, умножение матриц $4 × 4$ стандартным методом требует меньше арифметических опреаций, чем методом Штрассена. Метод Штрассена будет эффективнее работать на более больших матрицах.

**Ответ:** Для умножения матриц $4 × 4$ стандартным методом требуется $112$ арифметических операций, в то время как методом Штрассена $156$.

**9.Быстрое возведение в степень**

Объясните какие свойства квадратных матриц и операций над ними позволяют применить алгоритм быстрого возведения в степень (exponentiation by squaring). Например, вместо вычисления $R = (((A \cdot A)\cdot A)\cdot A)$ использовать вычисления $T = A \cdot A, R = T \cdot T$ с меньшим количеством матричных умножений.

**Решение:**

Алгоритм быстрого возведения в степень позволяет вычислять степень квадратной матрицы быстрее, благодаря рекурсивному подходу.

Применение алгоритма быстрого возведения в степень возможно благодаря ассоциативности умножения квадратных матриц, т.е. $(AB)C = A(BC)$ (результат не зависит от расстановки скобок). Например:

$$ R = (((A \cdot A)\cdot A)\cdot A) = (A \cdot A) \cdot (A \cdot A). $$

А также умножение квадратных матриц является замкнутой операцией, т.е. в результате умножения матриц $2 × 2$ получится также матрица $2 × 2$, что позволяет рекурсивно возводить получившийся результат в степень.
